```python
import random
import matplotlib.pyplot as plt
```

Loi des grands nombres

```python
x, y, mean, esperance = ([],[],[],[])

for i in range(10000):
    x.append(i)
    y.append(random.randint(1, 6))
    mean.append(sum(y)/len(y))
    esperance.append(3.5)

plt.plot(x,mean,'ro')
plt.plot(x,esperance)
plt.show()
```

Dés truquès

```python
x, y, mean = ([],[],[])

liste = [1,1,1,2,2,2,3,3,3,4,4,4,5,5,5,6]
for i in range(1000):
    x.append(i)
    y.append(random.choice(liste))
    mean.append(sum(y)/len(y))
    
plt.plot(x,mean,'ro')
plt.show()
```
